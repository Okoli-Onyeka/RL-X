```python
from semanticreasoning import ProcessData

```

```python
nlp = ProcessData("google/embeddinggemma-300m")
embeddings = nlp.getEmbeddings([
            "Move to the Left.",
            "Don't go right.",
            "Turn leftwards.",
            "Move to the right.",
            "Don't go left.",
            "Turn rightwards."
            ])
```

```python
similarities = nlp.getSimilarities(embeddings=embeddings)
```
