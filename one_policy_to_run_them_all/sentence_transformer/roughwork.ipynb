```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import torch

from neuralnetwork import NeuralNetwork
from semanticreasoning import ProcessData

pData = ProcessData()
text = pData.loadXData()
embeddings = pData.getEmbeddings(text)
y = pData.loadYData()
input_size = embeddings.shape[1]
output_size = y.shape[1]
```

```python
X = torch.tensor(embeddings)
y = torch.tensor(y)
```

```python
nn_model = NeuralNetwork(input_size, output_size)
nn_model.trainModel(X, y)
```

```python
nn_model.testModel(X, y)
```

```python
sentence = "chase to the left"

embedding = pData.getEmbeddings(sentence)

embedding = torch.tensor(
    embedding,
    dtype=torch.float32
).unsqueeze(0)   # make shape (1,input_size)

nn_model.model.eval()

with torch.no_grad():
    prediction = nn_model.model(embedding)

print(prediction)
```

```python
NeuralNetwork.saveModel(nn_model, "sentence_nn.pth")
```

```python
load_model = NeuralNetwork.loadModel(input_size, output_size, "sentence_nn.pth")

load_model.model.eval()

with torch.no_grad():
    prediction = load_model.model(embedding)
```
